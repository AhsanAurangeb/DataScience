```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_title('Set Title')
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

ax1.plot(x,y)
ax2.plot(y,x)
plt.show()
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2

fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

ax1.set_ylim(0,10000)
ax1.set_xlim(0,100)

ax2.set_xlim(20,22)
ax2.set_ylim(0,1)


ax1.set_xlabel('X Axis')
ax1.set_ylabel('Y Axis')
ax1.set_title('Set Title')

ax2.set_xlabel('X Axis')
ax2.set_ylabel('Y Axis')
ax2.set_title('Set Title')

ax1.plot(x,z)
ax2.plot(x,y)


plt.show()

```

```python
fig, ax = plt.subplots(1,2)


ax[0].plot(x,y,'--',lw=4,color='blue')
ax[1].plot(x,z,color='red',lw=4)
```

```python
fig, ax = plt.subplots(1,2,figsize=(12,2), dpi = 100)


ax[0].plot(x,y,lw=4,color='blue')
ax[1].plot(x,z,'--',color='red',lw=4)
```
